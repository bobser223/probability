### 3 Розподіл Пуассона

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин розподілу Пуассона з параметрами:
1. $\mu = 0.1$
2. $\mu = 0.2$
3. $\mu = 0.3$
4. $\mu = 0.4$
5. $\mu = 0.5$
6. $\mu = 0.6$
7. $\mu = 0.7$
8. $\mu = 0.8$
9. $\mu = 0.9$

```python
mus = np.arange(0.1, 1.0, 0.1)

for mu in mus:

    x = np.arange(0, int(mu*5)+1)
    pmf = poisson.pmf(x, mu)
    cdf = poisson.cdf(x, mu)
    sample = poisson.rvs(mu, size=10000)
    
    # PMF
    plt.figure(figsize=(8,5))
    plt.stem(x, pmf, basefmt=" ")
    plt.title(f'PMF Poisson(μ={mu:.1f})')
    plt.xlabel('k')
    plt.ylabel('P(X=k)')
    plt.grid(True)
    plt.show()
    
    # CDF
    plt.figure(figsize=(8,5))
    plt.step(x, cdf, where='post')
    plt.title(f'CDF Poisson(μ={mu:.1f})')
    plt.xlabel('k')
    plt.ylabel('F(X≤k)')
    plt.grid(True)
    plt.show()
    
    # Гістограма + теоретична PMF
    plt.figure(figsize=(8,5))
    plt.hist(sample, bins=x-0.5, density=True)
    plt.stem(x, pmf, basefmt=" ")
    plt.title(f'Histogram vs PMF Poisson(μ={mu:.1f})')
    plt.xlabel('k')
    plt.ylabel('Relative frequency')
    plt.grid(True)
    plt.show()

```

## 4 Апроксимація біноміального розподілу розподілом Пуассона

Покажіть схожість біноміального розподілу та розподілу Пуассона (їх функцій щільності) з відповідними параметрами.
Наприклад:
- $n = 100, p = 0.05$
- $n = 100, p = 0.1$
- $n = 100, p = 0.15$
- $n = 100, p = 0.2$
- $n = 100, p = 0.35$
- $n = 100, p = 0.5$
- $n = 100, p = 0.75$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson

n = 100
ps = [0.05, 0.1, 0.15, 0.2, 0.35, 0.5, 0.75]

for p in ps:
    lam = n * p
    x = np.arange(0, int(lam + 4*np.sqrt(lam)) + 1)
    
    pmf_bin = binom.pmf(x, n, p)
    pmf_poi = poisson.pmf(x, lam)
    
    plt.figure(figsize=(8,4))
    plt.stem(x, pmf_bin, basefmt=" ", label="Binomial(n=100,p={:.2f})".format(p))
    plt.stem(x, pmf_poi, basefmt=" ", linefmt='C1-', markerfmt='C1o', label="Poisson(λ={:.2f})".format(lam))
    plt.title(f"Binomial vs Poisson PMF (n=100, p={p:.2f}, λ={lam:.2f})")
    plt.xlabel("k")
    plt.ylabel("Probability")
    plt.legend()
    plt.grid(True)
    plt.show()


```
